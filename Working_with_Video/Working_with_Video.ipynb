# 4. Working with Video Captures

```python
# Establish capture
cap = cv2.VideoCapture(video_path)
# Loop through each frame
for frame_idx in range(int(cap.get(cv2.CAP_PROP_FRAME_COUNT))):
    
    # Read frame 
    ret, frame = cap.read()
    
    # Gray transform
    XYZ = cv2.cvtColor(frame, cv2.COLOR_BGR2XYZ)
    
    # Show image
    cv2.imshow('Video Player', XYZ)
    
    # Breaking out of the loop
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break

# Close down everything
cap.release()
cv2.destroyAllWindows()
```

# 5.Writing out Videos

```python
# Establish capture
cap = cv2.VideoCapture(video_path)

# Properties
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
fps = cap.get(cv2.CAP_PROP_FPS)

# Video Writer 
video_writer = cv2.VideoWriter(os.path.join('images_and_videos', 'Videos', 'kaikai_kitan_piano_XYZ.mp4'),
                               cv2.VideoWriter_fourcc('P','I','M','1'), fps, (width, height), isColor=True) 

# In the gray video case, isColor should be false or it will be an error.
                                                                                        
# Loop through each frame
for frame_idx in range(int(cap.get(cv2.CAP_PROP_FRAME_COUNT))):
    
    # Read frame 
    ret, frame = cap.read()
    
    # Gray transform
    XYZ = cv2.cvtColor(frame, cv2.COLOR_BGR2XYZ)
    
    # Show image
    cv2.imshow('Video Player', XYZ)
    
    # Write out frame 
    video_writer.write(XYZ)
    
    # Breaking out of the loop
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break

# Close down everything
cap.release()
cv2.destroyAllWindows()
# Release video writer
video_writer.release()
```
