# 1.Read in Image

```python
img_path = os.path.join('images_and_videos', 'images', 'Dog_and_Me.png')
```

```python
img = cv2.imread(img_path) # return numy array with bgr type
```

```python
print(img)
```

# 2.Render Image

```python
from matplotlib import pyplot as plt
```

```python
# Show image
plt.imshow(img)
```

```python
recolor_to_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(recolor_to_rgb)
plt.show() # Remove the properties feature at the first line.
```

```python
recolor = cv2.cvtColor(img, cv2.COLOR_BGR2XYZ) # you can try Lab , XYZ ... etc
plt.imshow(recolor)
plt.show() # Remove the properties feature at the first line.
```

# 3. Write out Image

```python
output_path = os.path.join('images_and_videos', 'images', 'Dog_and_Me_XYZ.png')
```

```python
cv2.imwrite(output_path,  recolor)
```
